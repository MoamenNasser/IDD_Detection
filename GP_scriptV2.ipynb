```python
#needs to be rewritten
def listdirs(rootdir):
    for it in os.scandir(rootdir):
        if it.is_dir():
            list_dir.append(it.path)
            listdirs(it)
def get_dir(rootdir):
  global list_dir
  list_dir=[]
  listdirs(rootdir)
  print(len(list_dir))
  max=0
  for i in list_dir:
    depth=i.count('/')
    if(depth>max):
      max=depth
  for i in list_dir:
    if(i.count('/') != max):
      list_dir.remove(i)
  print(len(list_dir))
  return list_dir
```

```python
list_dir=get_dir(paths['annotaions'])
```

```python
counter=0
absent_annotations_folders=[]
for dir in os.listdir(paths['images']):
  dir=os.path.join(paths['images'],dir)
  for folder in os.listdir(dir):
    folder=os.path.join(dir,folder)
    for image in os.listdir(folder):
      image_dir=os.path.join(folder,image)
      if(image_dir.endswith('.jpg')):
        annotation_dir=image_dir.replace('JPEGImages','Annotations')
        annotation_dir=annotation_dir.replace('.jpg','.xml')
        if(not os.path.exists(annotation_dir)):
          print(image_dir+' annotations does not exist')
          basename=os.path.basename(annotation_dir)
          annotation_dir=annotation_dir[:-len(basename)]
          if( annotation_dir not in absent_annotations_folders):
            absent_annotations_folders.append(annotation_dir)
          counter+=1
      else:
        print(image_dir)+' this is a folder'
for i in range(len(absent_annotations_folders)):
  start=absent_annotations_folders[i].find('Annotations')
  start=start+len('Annotations/') 
  absent_annotations_folders[i]=absent_annotations_folders[i][start:-1]
```

```python
def delete_absent_from_text(text_dir):
  basename=os.path.basename(text_dir)
  print(basename)
  new_lines=0
  File=open(text_dir,'r')
  lines=File.readlines()
  File.close()
  print('Total lines: {}'.format(len(lines)))

  File=open(text_dir[:-4]+'_new.txt','w')
  for line in lines:
    basename=os.path.basename(line)
    line_stripped=line[:-len(basename)-1]
    if line_stripped not in absent_annotations_folders:
      File.write(line)
      new_lines+=1
  File.close()
  print('Total lines after removal {}'.format(new_lines))
```

```python
delete_absent_from_text(paths['train_txt'])
delete_absent_from_text(paths['test_txt'])
delete_absent_from_text(paths['val_txt'])
```

# 5.Converting to YOLO format:

```python
def pascal_voc_to_yolo():
  for folder in list_dir:
    dir=folder
    images_dir=dir.replace('Annotations','JPEGImages')
    basename=os.path.basename(folder)
    print(folder+' in progress')
    ##dir=os.path.join('/content/data/IDD_Detection/Annotations',folder)
    #images_dir=os.path.join('/content/data/IDD_Detection/JPEGImages',folder)
    shutil.copy2(paths['script'],dir)
    os.chdir(dir)
    subprocess.call("find . -name '*.xml' -exec python conv.py {} \;", shell=True)
    for file in os.listdir():
      if(file.endswith('.txt') and not os.path.exists(os.path.join(images_dir,file)) ):
        shutil.move(file, images_dir)
    os.remove('conv.py')
    print('Folder Done')
```
